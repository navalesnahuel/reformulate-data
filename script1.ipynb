```python
import pandas as pd
```

```python
data_import = pd.read_excel(r"C:\Users\PC 2\Desktop\my_projects\data_reformulate\Data - Survey Monkey Output Edited.xlsx", sheet_name='Edited_Data')
data_import
```

```python
data_import.columns
to_drop = [ 'Start Date', 'End Date', 'Email Address ',
       'First Name ', 'Last Name ', 'Custom Data 1 ']
data_import.drop(columns=to_drop, inplace=True)
```

```python
data_melted = data_import.copy()
id_vars = list(data_melted.columns)[:8]
value_vars = list(data_melted.columns)[8:]
```

```python
data_melted = pd.melt(data_melted, id_vars=id_vars, value_vars=value_vars, var_name='Question + Subquestion', value_name='Answer')
```

```python
data_melted
```

```python
questions = pd.read_excel(r"C:\Users\PC 2\Desktop\my_projects\data_reformulate\Data - Survey Monkey Output Edited.xlsx", sheet_name='Question')
questions.drop(columns=['Raw_Question', 'Raw_Subquestion', 'Subquestion'], inplace=True)
questions
```

```python
data_merge = data_melted.copy()
```

```python
data_merge = data_merge.merge(questions, how='left', left_on='Question + Subquestion', right_on='Question + Subquestion')
```

```python
data_cleaned = data_merge.copy()
```

```python
non_null_values = data_cleaned[data_cleaned['Answer'].notna()]
count = non_null_values.groupby('Question')['Respondent ID'].nunique().reset_index()
count.rename(columns={'Respondent ID':'Total_Respondents'}, inplace=True)
count
```

```python
data_merge2 = data_cleaned.merge(count, how='left', left_on='Question', right_on='Question')
data_merge2
```

```python
data_third_merge = data_merge2.copy()
```

```python
# data_third_merge
same_answer = data_merge2.groupby(['Question + Subquestion','Answer'])['Respondent ID'].nunique().reset_index()
same_answer.rename(columns={'Respondent ID':'Same_Answer'}, inplace=True)
same_answer
```

```python
data_third_merge = data_third_merge.merge(same_answer, how='left', left_on=['Question + Subquestion', 'Answer'], right_on=['Question + Subquestion', 'Answer'])
```

```python
data_third_merge['Same_Answer'].fillna(0, inplace=True)
```

```python
data_third_merge
```

```python
data_third_merge.rename(columns={'Identify which division you work in. - Response' : 'Primary_Division',
                                 	'Identify which division you work in. - Other (please specify)': 'Secondary_Division',
                                    'Which of the following best describes your position level? - Response': 'Position_Level',	
                                    'Which generation are you apart of? - Response': 'Generation',
                                 'Please select the gender in which you identify. - Response': 'Gender',	
                                 'Which duration range best aligns with your tenure at your company? - Response':'Tenure',
                                     	'Which of the following best describes your employment type? - Response':'Employment_Type'},inplace=True)
```

```python
output = data_third_merge.copy()
```
